# Concatenating
- Concatenate DataFrames along row and column.

```python
import pandas as pd
```

```python
data1 = {
        'id': ['1', '2', '3', '4', '5'],
        'Feature1': ['A', 'C', 'E', 'G', 'I'],
        'Feature2': ['B', 'D', 'F', 'H', 'J']}

data2 = {
        'id': ['1', '2', '6', '7', '8'],
        'Feature1': ['K', 'M', 'O', 'Q', 'S'],
        'Feature2': ['L', 'N', 'P', 'R', 'T']}
```

```python
df1 = pd.DataFrame(data1, columns = ['id', 'Feature1', 'Feature2'])
df1
```

```python
df2 = pd.DataFrame(data2, columns = ['id', 'Feature1', 'Feature2'])
df2
```

```python
df_row = pd.concat([df1, df2])
df_row
```

```python
df_col = pd.concat([df1,df2], axis=1)

df_col
```

# Merging
- Merge DataFrames on specific keys by different join logics like left-join, inner-join etc.

```python
data3 = {
        'id': ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],
        'Feature3': [12, 13, 14, 15, 16, 17, 15, 12, 13, 23]}
```

```python
df3 = pd.DataFrame(data3, columns = ['id', 'Feature3'])
df3
```

```python
df_merge_col = pd.merge(df_row, df3, on='id')

df_merge_col
```

# Joining
- Join DataFrames by index

```python
df2.join(df3, lsuffix='_left', rsuffix='_right')
```

```python
df2.join(df3.set_index('id'), on='id')
```
