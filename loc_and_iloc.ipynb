# loc() and iloc()
- both used in filtering the data according to some conditions

### loc() : label based data selecting method
- pass the name of the row or column
- include the last element of the range passed in it
- accept the boolean data

```python
import pandas as pd
 
data = pd.DataFrame({'Brand': ['Maruti', 'Hyundai', 'Tata',
                               'Mahindra', 'Maruti', 'Hyundai',
                               'Renault', 'Tata', 'Maruti'],
                     'Year': [2012, 2014, 2011, 2015, 2012,
                              2016, 2014, 2018, 2019],
                     'Kms Driven': [50000, 30000, 60000,
                                    25000, 10000, 46000,
                                    31000, 15000, 12000],
                     'City': ['Gurgaon', 'Delhi', 'Mumbai',
                              'Delhi', 'Mumbai', 'Delhi',
                              'Mumbai', 'Chennai',  'Ghaziabad'],
                     'Mileage':  [28, 27, 25, 26, 28,
                                  29, 24, 21, 24]})
display(data)
```

```python
# selecting with condition
data.loc[data['Brand']=='Maruti']
```

```python
# selecting range of rows
data.loc[2:5]
```

```python
# updating values
data.loc[(data['Year']<2015), ['Mileage']] = 22
data
```

### iloc(): index based data selecting method
- pass an integer index
- does not include last element of the range passed in it
- does not accept the boolean data

```python
# selecting rows with index
data.iloc[[0, 2, 4, 7]]
```

```python
# selecting a range of columns and rows
data.iloc[1:5, 2:5]
```
